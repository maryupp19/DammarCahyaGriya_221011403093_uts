```python
# List of company details
company_detail_list = [
    {
        'name': 'Company 1',
        'domain': 'Retail',
        'country': 'United States'
    },
    {
        'name': 'Company 2',
        'domain': 'Technology',
        'country': 'United Kingdom'
    },
    {
        'name': 'Company 3',
        'domain': 'Healthcare',
        'country': 'United States'
    }
]

# List of employee details
employee_detail_list = [
    {
        'name': 'EMP-0001',
        'first_name': 'John',
        'last_name': 'Doe',
        'full_name': 'John Doe',
        'company': 'Company 1',
        'nationality': 'Australia'
    },
    {
        'name': 'EMP-0002',
        'first_name': 'Tom',
        'last_name': 'Smith',
        'full_name': 'Tom Smith',
        'company': 'Company 2',
        'nationality': 'United States'
    },
    {
        'name': 'EMP-0003',
        'first_name': 'Andrew',
        'last_name': 'Sebastian',
        'full_name': 'Andrew Sebastian',
        'company': 'Company 3',
        'nationality': 'United States'
    },
    {
        'name': 'EMP-0005',
        'first_name': 'Ying Han',
        'last_name': 'Tan',
        'full_name': 'Ying Han Tan',
        'company': 'Company 1',
        'nationality': 'Australia'
    },
    {
        'name': 'EMP-0015',
        'first_name': 'Kenneth',
        'last_name': 'Ng',
        'full_name': 'Kenneth Ng',
        'company': 'Company 3',
        'nationality': 'United States'
    },
    {
        'name': 'EMP-0018',
        'first_name': 'Rubby',
        'last_name': 'Lee',
        'full_name': 'Rubby Lee',
        'company': 'Company 2',
        'nationality': 'Hong Kong'
    },
    {
        'name': 'EMP-0017',
        'first_name': 'Robert',
        'last_name': 'White',
        'full_name': 'Robert White',
        'company': 'Company 1',
        'nationality': 'United Kingdom'
    }
]

def get_companies_with_employee_nationalities(companies, employees):
    # Create a dictionary to hold the result
    company_nationality_map = {company['name']: {} for company in companies}
    
    for employee in employees:
        company_name = employee['company']
        nationality = employee['nationality']
        if nationality not in company_nationality_map[company_name]:
            company_nationality_map[company_name][nationality] = 0
        company_nationality_map[company_name][nationality] += 1
    
    # Convert the dictionary to the expected format
    result = []
    for company in companies:
        result.append({
            'company': company['name'],
            'employee_nationality': company_nationality_map[company['name']]
        })
    
    return result

# Get the companies with the number of employees' nationalities
result = get_companies_with_employee_nationalities(company_detail_list, employee_detail_list)

# Print the result
print(result)
```
